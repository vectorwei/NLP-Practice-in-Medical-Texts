第一步：去除大写和标点和数字

```python
import re
file1 = open('new.txt', 'r', encoding='utf-8') # 打开要去掉空行的文件
file2 = open('new1.txt', 'w', encoding='utf-8') # 生成没有空行的文件

for text in file1.readlines():    
    text=text.lower() 
    text = re.sub(r'[^a-zA-Z]', " ", text)
    file2.write(text)
print('输出成功....')  
file1.close()
file2.close()
```
